```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import sklearn
```

```python
inputs = np.linspace(-1,1,5)
```

```python
target = inputs**2 
```

```python
def reservoir(u):
    return np.array([inputs,inputs**2,np.sin(3*inputs),np.cos(5*inputs)])
```

```python
phi = reservoir(inputs)
```

```python
W = target @ np.linalg.pinv(phi)
```

```python
inputs_test = np.linspace(-1,1,20)
```

```python
phi_test = reservoir(inputs_test)
```

```python
target_pred = W @ phi_test
```

```python
plt.plot(inputs_test,target_pred,label="True")
```
