```python
# Defining a class
class Dog:
    # Constructor method (runs when creating an object)
    def __init__(self, name, age):
        self.name = name   # attribute
        self.age = age     # attribute

    # Method (function inside class)
    def bark(self):
        print(f"{self.name} is barking!")

# Creating objects (instances of Dog)
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 5)

# Accessing attributes
print(dog1.name)  # Buddy
print(dog2.age)   # 5

# Calling a method
dog1.bark()  # Buddy is barking!
```

```python
class Circle:
    def __init__(self,radius,centre):
        self.radius = radius
        self.centre = centre
    
    def equation_of_circle(self):
        #print(f"equation of circle is {self.centre}")
        print(f"equation of circle is {self.radius}^2 = (x - {self.centre[0]})^2 + (y - {self.centre[1]})^2")
        
```

```python
# Example: Using ML for system identification in control systems

# Assume you have input-output data from a control system
# X: input signal (e.g., voltage), y: output signal (e.g., position)

# Generate synthetic data for demonstration
import numpy as np 

time = np.linspace(0, 10, 100)
X = np.sin(time)  # input signal
y = 2 * X + np.random.normal(0, 0.1, size=time.shape)  # output with noise

# Split data for training/testing
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1,1), y, test_size=0.2, random_state=42)

# Train a regression model to identify system behavior
reg = LinearRegression()
reg.fit(X_train, y_train)

# Predict output for test input
y_pred = reg.predict(X_test)

# Plot actual vs predicted output
plt.scatter(X_test, y_test, label='Actual')
plt.scatter(X_test, y_pred, label='Predicted', color='red')
plt.xlabel('Input Signal')
plt.ylabel('Output Signal')
plt.legend()
plt.title('ML System Identification Example')
plt.show()
```
